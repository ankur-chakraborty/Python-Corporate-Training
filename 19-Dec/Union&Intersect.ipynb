```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import when, col

spark = SparkSession.builder.appName("SET").getOrCreate()
```

```python
data_a=[
    ("U1","Python"),
    ("U2","Java"),
    ("U3","Spark")


]

df_a=spark.createDataFrame(data_a,["user_id","course"])
```

```python
data_b=[
    ("U2","Java"),
    ("U3","Spark"),
    ("U4","Python")
]

df_b=spark.createDataFrame(data_b,["user_id","course"])
```

```python
df_a.union(df_b).show()
```

```python
df_a.intersect(df_b).show() #Expensive operation (USES LOT OF RAM & DISK)
```
