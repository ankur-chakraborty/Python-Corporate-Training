Ex1

```python
rdd=spark.sparkContext.textFile("telecom.csv")
rdd.take(5)
```

Ex2

```python
header = rdd.first()
data_rdd=rdd.filter(lambda row: row !=header)
data_rdd.collect()
```

Ex3

```python
split_rdd=data_rdd.map(lambda row: row.split(","))
split_rdd.collect()
```

Ex4

```python
total_cost=split_rdd.map(lambda x: (x[3], float(x[6])))
total_cost.collect()
```

```python
cost_city=total_cost.reduceByKey(lambda x,y: x+y)
cost_city.collect()
```

Ex5

```python
max_cost=cost_city.reduce(
    lambda x,y: x if x[1]>y[1] else y
)
max_cost
```

Ex6

```python

duration_by_type = (
    split_rdd
    .map(lambda x: (x[4], int(x[5])))
    .reduceByKey(lambda a, b: a + b)
)
duration_by_type.collect()
```

Ex7

```python

calls_per_city = (
    split_rdd
    .map(lambda x: (x[3], 1))              # (city, 1)
    .reduceByKey(lambda a, b: a + b)       # sum counts per city
)

calls_per_city.collect()
```

Ex8


```python
# Map to (city, (cost, 1)) with explicit float conversion
city_cost_count = split_rdd.map(lambda x: (x[3], (float(x[6]), 1)))

# Reduce to (sum_cost, count) per city
city_sum_count = city_cost_count.reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1]))

# Compute average = sum_cost / count
avg_cost_per_city = city_sum_count.mapValues(lambda sc: sc[0] / sc[1])

avg_cost_per_city.collect()
```

Ex9

```python
high_value_calls = split_rdd.filter(lambda x: float(x[6]) > 20.0)

high_value_calls.collect()
```

Ex10

```python
isd_calls_per_city = (
    split_rdd
    .filter(lambda x: x[4].upper() == "ISD")   # keep only ISD calls
    .map(lambda x: (x[3], 1))                  # (city, 1)
    .reduceByKey(lambda a, b: a + b)           # count per city
)

isd_calls_per_city.collect()
```

Ex11

```python
longest_call = split_rdd.max(key=lambda x: int(x[5]))
longest_call
```

Ex12

```python
revenue_by_caller = (
    split_rdd
    .map(lambda x: (x[1], float(x[6])))   # (caller, cost)
    .reduceByKey(lambda a, b: a + b)      # sum costs per caller
)

revenue_by_caller.collect()
```

Ex13

```python
suspicious_calls = (
    split_rdd
    .filter(lambda x: int(x[5]) > 900 and float(x[6]) > 25.0)
)

suspicious_calls.collect()
```
