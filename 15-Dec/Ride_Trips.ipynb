#ANALYTICS THINKING QUESTIONS

Exercise 15

Which city generates the highest total trip_fare?

```python
from pyspark.sql.functions import sum as spark_sum
city_revenue_df = df.groupBy("city").agg(spark_sum("trip_fare").alias("total_revenue")) \
.orderBy("total_revenue", ascending = False)
city_revenue_df.show(1)
```

Exercise 16

Which vehicle_type has the highest average fare?

```python
from pyspark.sql.functions import avg
vehicle_type_avg_fare_df = df.groupBy("vehicle_type").agg(avg("trip_fare").alias("avg_fare")) \
.orderBy("avg_fare", ascending = False)
vehicle_type_avg_fare_df.show(1)
```

Exercise 17

Which driver has completed the most trips?

```python
from pyspark.sql.functions import count
driver_trips_df = df.groupBy("driver_name").agg(count("*").alias("trip_count")) \
.orderBy("trip_count", ascending = False)
driver_trips_df.show(1)
```

Exercise 18

Why is Parquet preferred for analytics dashboards and aggregations?

```python
csv_df = spark.read \
.option("header", True) \
.option("inferSchema", True) \
.csv("/content/trips.csv")

csv_agg = csv_df.groupBy("city") \
.sum("trip_fare") \

csv_agg.explain(True)
```

Challenge 2
Create a summary dataset with:
city
total_trips
total_revenue
average_trip_duration
Write it to Parquet.

```python
from pyspark.sql.functions import sum as spark_sum, avg, count
summary_df = df.groupBy("city") \
.agg(
    count("*").alias("total_trips"),
    spark_sum("trip_fare").alias("total_revenue"),
    avg("trip_duration_minutes").alias("average_trip_duration")
)
summary_df.write \
.mode("overwrite") \
.parquet("summary.parquet")
```
