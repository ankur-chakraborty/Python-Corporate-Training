#3 SPECIFIC COLUMNS

```python
df.select("customer_id","name","annual_spend").show(30)
```

#more than 60k

```python
df.filter(df.annual_spend>60000).show()
```

#From delhi and <30

```python
# df2=df.filter(df.city=="Delhi")
# df2.filter(df.age<30).show()

df.filter((df.city == "Delhi") & (df.age < 30)).show()
```

#Spend lakh

```python
from pyspark.sql.functions import col
df2=df.withColumn("Spend_lakh",col("annual_spend")/100000)
df2.show()
```

new col

```python
from pyspark.sql.functions import when
df = df.withColumn(
    "customer_type",
    when(df.annual_spend > 70000, "Premium").otherwise("Standard")
)
df.show()
```

starts with A

```python
df.filter(df.name.startswith("A")).show()
```

#Category clothing or electronics

```python
df.filter((df.category == "Clothing") | (df.category == "Electronics")).show()
```

Upper

```python
from pyspark.sql.functions import upper
df2=df.withColumn("name_upper", upper(col("city")))
df2.show()
```

#Remove Category

```python
df.drop("category").show()
```

Sort cust by age desc

```python
df.orderBy(df.age.desc()).show()
```

filter age

```python

from pyspark.sql.functions import avg, col


avg_age = df.select(avg(col("age")).alias("avg_age")).collect()[0]["avg_age"]

df.filter(col("age") < avg_age).show()
```
